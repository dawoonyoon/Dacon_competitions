## üëã Ïã§Ïäµ
---
1. train Îç∞Ïù¥ÌÑ∞ Î≥ÄÌôò
2. Î™®Îç∏ ÌïôÏäµ (Decision Tree)
3. test Îç∞Ïù¥ÌÑ∞ Î≥ÄÌôò

```python
# ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∞è Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
import pandas as pd

# PolynomialFeatures ÎùºÏù¥Î∏åÎü¨Î¶¨ Ìò∏Ï∂ú
from sklearn.preprocessing import PolynomialFeatures

# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

# Î≤îÏ£ºÌòï Î≥ÄÏàò Ï†úÍ±∞
train.drop('type', axis = 1,inplace = True)
test.drop('type', axis = 1,inplace = True)

# index Î≥ÄÏàò Ï†úÍ±∞

train.drop('index', axis = 1,inplace = True)
test.drop('index', axis = 1,inplace = True)
```

```python
# Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏
train.head()
```

```python
 # train Îç∞Ïù¥ÌÑ∞Î•º PolynomialFeatures Î•º Ïù¥Ïö©ÌïòÏó¨ Î≥ÄÌôò
poly_features = PolynomialFeatures(degree=2)
df = train.drop('quality', axis=1)
train_poly = poly_features.fit_transform(df)
train_poly = pd.DataFrame(train_poly)
train_poly.head()
```

```python
# DecisionTreeClassifier Î™®Îç∏ÏùÑ Î≥ÄÌôòÎêú train Îç∞Ïù¥ÌÑ∞Î°ú ÌïôÏäµ

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(train_poly, train['quality'])
```

```python
# test Îç∞Ïù¥ÌÑ∞ Î≥ÄÌôò
poly_features = PolynomialFeatures(degree=2)
test_poly = poly_features.fit_transform(test)
test_poly = pd.DataFrame(test_poly)
test_poly
```

```python
# Í≤∞Í¥èÍ∞í Ï∂îÎ°†
pred = model.predict(test_poly)
```

```python
# Ï†ïÎãµ ÌååÏùº ÏÉùÏÑ±
submission = pd.read_csv("data/sample_submission.csv")
submission['quality'] = pred
submission.to_csv('poly.csv', index=False)
```
